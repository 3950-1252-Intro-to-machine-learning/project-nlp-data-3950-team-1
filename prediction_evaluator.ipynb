```python
import pandas as pd
import os

def scoreChecker(df_sub, df_correct):
    """
    Standardized evaluator to compare your Grid Search results against the key.
    """
    # Ensure IDs are strings to prevent merge mismatches
    df_sub['id'] = df_sub['id'].astype(str)
    df_correct['id'] = df_correct['id'].astype(str)
    
    # Force suffixes _x (prediction) and _y (actual) as per project requirements
    check_full = pd.merge(df_sub, df_correct, how='inner', on='id', suffixes=('_x', '_y'))
    
    if check_full.empty:
        print("Error: No matching IDs found between submission and key.")
        return 0, 0, check_full

    # Column mappings for the 6 toxicity categories
    corr_cols = ["toxic_y", "severe_toxic_y", "obscene_y", "threat_y", "insult_y", "identity_hate_y"]
    sub_cols = ["toxic_x", "severe_toxic_x", "obscene_x", "threat_x", "insult_x", "identity_hate_x"]
    
    correct = 0
    total = 0
    
    for index, sub_row in check_full.iterrows():
        # Only score rows where the key has valid data (>= 0)
        if sub_row["toxic_y"] >= 0:
            total += len(corr_cols)
            for i in range(len(corr_cols)):
                if sub_row[sub_cols[i]] == sub_row[corr_cols[i]]:
                    correct += 1
                    
    return correct, total, check_full
```

```python
import pandas as pd
import os

def scoreChecker(df_sub, df_correct):
    # 1. Clean column names to remove any existing _x or _y before merging
    # This prevents 'toxic_x_x' or other naming confusion
    df_sub.columns = [c.replace('_x', '').replace('_y', '') for c in df_sub.columns]
    df_correct.columns = [c.replace('_x', '').replace('_y', '') for c in df_correct.columns]

    # 2. Standardize IDs
    df_sub['id'] = df_sub['id'].astype(str)
    df_correct['id'] = df_correct['id'].astype(str)
    
    # 3. FORCE the suffixes during merge
    check_full = pd.merge(df_sub, df_correct, how='inner', on='id', suffixes=('_x', '_y'))
    
    if check_full.empty:
        print("Error: IDs in out.csv do not match IDs in sample_correct.csv")
        return 0, 0, check_full

    # 4. Define the exact column names the PDF logic looks for
    corr_cols = ["toxic_y", "severe_toxic_y", "obscene_y", "threat_y", "insult_y", "identity_hate_y"]
    sub_cols = ["toxic_x", "severe_toxic_x", "obscene_x", "threat_x", "insult_x", "identity_hate_x"]
    
    correct = 0
    total = 0
    
    # 5. Calculate scores based on the instructor's logic
    for index, sub_row in check_full.iterrows():
        if sub_row["toxic_y"] >= 0:
            total += len(corr_cols)
            for i in range(len(corr_cols)):
                if sub_row[sub_cols[i]] == sub_row[corr_cols[i]]:
                    correct += 1
                    
    return correct, total, check_full

# --- Execution Section ---
sub_path = "output/out.csv"
key_path = "data/sample_correct.csv"

if os.path.exists(sub_path) and os.path.exists(key_path):
    submission = pd.read_csv(sub_path)
    key = pd.read_csv(key_path)
    
    correct, total, df_res = scoreChecker(submission, key)
    
    if total > 0:
        print(f"{correct} {total} {(correct*100/total):.1f}")
    else:
        print("No valid data found to evaluate.")
else:
    print("One or more files are missing. Check /output and /data folders.")
```
